#### **4. Clean and tokenize text**

```python
import re

#re.sub(pattern, replacement, string)

def remove_retweets_mentions(text):
    # Remove retweets
    text = str(text)
    text = re.sub(r'RT\s*@\w+:?', '', text)
    # Remove mentions
    text = re.sub(r'@\w+', '', text)
    return text

def remove_hyperlinks(text):
    # Define a regular expression pattern to match URLs
    url_pattern = r'https?://\S+|www\.\S+'

    # Use the sub() function to replace matched URLs with an empty string
    cleaned_text = re.sub(url_pattern, '', text)

    return cleaned_text #remove the hyperlinks

def remove_special_characters(text):
    # Define a pattern to match special characters (excluding underscores and whitespace)
    pattern = r'[^a-zA-Z0-9\s_]'

    # Use the sub() function to replace matched special characters with an empty string
    cleaned_text = re.sub(pattern, '', text)

    return cleaned_text #remove special characters

twitter_en['cleaned_text'] = twitter_en['tweet_text'].apply(remove_retweets_mentions)
twitter_en['cleaned_text'] = twitter_en['cleaned_text'].apply(remove_hyperlinks)
twitter_en['cleaned_text'] = twitter_en['cleaned_text'].apply(remove_special_characters)
twitter_en['cleaned_text'] = twitter_en['cleaned_text'].str.lower()
twitter_en['cleaned_text'].head(5)
```

```python
#tokenize tweets
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
tokens = []

for text in twitter_en.cleaned_text:
    words = word_tokenize(str(text))
    for word in words:
        tokens.append(word) #tokenize the words in the document and append them to a list
```

```python
tokens = [token for token in tokens if token.strip() != ""] #remove blank space
tokens = [item for item in tokens if len(item) > 1] #filters out any single-character words such as "a", "I", punctuation marks, or other tokens that are not meaningful by themselves in the given context.
# Get the stopwords once and store them in a set for faster lookup
stop_words = set(stopwords.words('english'))
# Remove stopwords from tokens
tokens = [word for word in tokens if word not in stop_words]
```



#### **5. Religion Dictionary**

##### **5.1. Open Dictionary**

```python
#Open religion dictionary
rels = pd.read_csv(r'/content/Religion Dictionary(Sheet1) (1).csv') #open the dictionary
print(rels.head(10)) #check first 10 rows
```

```python
#Create a function to lookup and assign categories
def assign_category(text):
    matching_categories = rels[rels['word'].apply(lambda word: str(word) in text)] #this is a function telling that for words in text, see if each word matches with word in the dictionary

    if not matching_categories.empty: #when there is a match
        return ', '.join(matching_categories['category']) #list the category name. When there are more than two matches, put category names joined by ','
    else:
        return '' # if there is no match, return ''
```

##### **5.2. Categorize tweets**

```python
twitter_en['category'] = twitter_en['cleaned_text'].apply(assign_category) ## Apply to tweet_text and make another column, 'category' containing list of categories associated with text
```

```python
twitter_en[['tweet_text','category']].head(5) #check the first row to see how categories are associated with each text
twitter_en['category'].value_counts()
```

```python
#Split the 'category' column by comma and create a new DataFrame with each category in a separate row
categories_split = twitter_en['category'].str.split(', ').explode().reset_index(drop=True) #explode to check all the categories associated with the dataset
unique_categories = categories_split.unique() #only retain unique categories
print(unique_categories) #print unique categories found in the dataset
```

```python
twitter_en['category'] = twitter_en['category'].astype(str) #make sure the category column is in string
categories_split = twitter_en['category'].str.split(', ') # Split the 'category' column by comma (', ') and create binary columns
print(categories_split) #look at the results

for new_column in unique_categories:
    twitter_en[new_column + '_rels'] = categories_split.apply(lambda x: x.count(new_column)) #create columns and count the number of the category in the list

twitter_en.fillna(0, inplace=True) # Fill NaN values with 0 in the binary columns

twitter_en.drop(columns=['' + '_rels'], inplace=True) #drop column ''
```

#### **7. NRC Dictionary**

```python
#Apply NRC dictionary to tweets
!pip install NRCLex

from nrclex import NRCLex #import libraries needed from the package
import nltk #get base data for NRC dictionary
```

```python
#Function to calculate emotion frequency in text
def calculate_emotion_frequency(text):
    # Create NRCLex object for the text
    text_lex = NRCLex(text)

    # Get emotions and their frequencies
    emotion_frequency = text_lex.raw_emotion_scores

    return emotion_frequency

twitter_en['NRC_category'] = twitter_en['cleaned_text'].apply(calculate_emotion_frequency)
```

```python
twitter_en[['tweet_text','NRC_category']].head(5) #check the first row to see how categories are associated with each text
```

```python
#Initialize a list to hold the unique emotion words
unique_emotions = set()

#Extract all unique emotions from the NRC_lexicon dictionaries
for lexicon in twitter_en['NRC_category']:
    unique_emotions.update(lexicon.keys())

unique_emotions

#Create columns for each unique emotion
for emotion in unique_emotions:
    twitter_en[emotion + '_NRC'] = twitter_en['NRC_category'].apply(lambda x: x.get(emotion, 0))

#Drop the original 'NRC_lexicon' column
twitter_en.drop(columns=['NRC_category'], inplace=True)

print(twitter_en)
```
